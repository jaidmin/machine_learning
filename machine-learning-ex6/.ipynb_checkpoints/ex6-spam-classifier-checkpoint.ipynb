```python
import string
import re
from nltk.stem import *

def process_email(email_path):
    sample1 = open(email_path)
    sample1 = sample1.read()


    sample1 = sample1.lower()


    stemmer = PorterStemmer()

    pattern_dollar = re.compile('[$]')
    sample1 = re.sub(pattern_dollar,'dollar',sample1)
    pattern_mail = re.compile('[\w\.-]+@[\w\.-]+')
    sample1 = re.sub(pattern_mail, 'emailaddr', sample1)
    pattern_url = re.compile('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    sample1 = re.sub(pattern_url,'httpaddr',sample1)
    pattern_alphanum = re.compile('\W')
    sample1 = re.sub(pattern_alphanum, ' ', sample1)
    pattern_spaces = re.compile(' +')
    sample1 = re.sub(pattern_spaces, ' ', sample1)
    pattern_numbers = re.compile('\d+')
    sample1 = re.sub(pattern_numbers, 'number', sample1)
    sample1 = sample1.strip()



    tst = sample1.split()
    singles = [stemmer.stem(plural) for plural in tst]
    vocabs = np.genfromtxt('./ex6/vocab.txt', dtype=str)
    vocabs = vocabs[:,1]
    vocabs = vocabs.tolist()
    return [(vocabs.index(item)+1) for item in singles if item in vocabs ]
```

```python
def create_feature_vector(email):
    words_in_email = process_email(email)
    features = np.zeros((1899,1))
    
    for item in words_in_email:
        features[item-1] =1
    
    return features
    
    


    
```
